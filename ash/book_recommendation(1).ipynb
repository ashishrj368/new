```python
import pandas as pd
import numpy as np
book_df = pd.read_csv('Books.csv')
book_df.head()
```

```python
user_df = pd.read_csv('Users.csv')
user_df.head()
```

```python
rating_df = pd.read_csv('Ratings.csv').sample(4000)
rating_df.head()
```

```python
user_rating = rating_df.merge(user_df,left_on='User-ID',right_on='User-ID')
user_rating.head()
```

```python
book_user_rating = user_rating.merge(book_df,left_on='ISBN',right_on='ISBN')
book_user_rating.head()
```

```python
book_user_rating = book_user_rating[['ISBN','Book-Title','Book-Author','User-ID','Book-Rating']]
book_user_rating.reset_index(drop=True,inplace=True)
book_user_rating.head()
```

```python
book_user_rating.isna().sum()
```

```python
df=book_user_rating[book_user_rating['Book-Rating']  >=5]
df.head()
```

```python
max_rating = book_user_rating["Book-Rating"].max()
min_rating = book_user_rating['Book-Rating'].min()
book_user_rating['Book-Rating'] = (book_user_rating['Book-Rating']-min_rating)/(max_rating-min_rating)*10
```

```python
book_user_rating['Book-Rating']
```

```python
user_item_matrix=book_user_rating.pivot_table(index='User-ID',columns='ISBN',values='Book-Rating',fill_value=0)
user_item_matrix.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_similarity = cosine_similarity(user_item_matrix)
user_similarity
```

```python
def recommend(user_id,user_similarity,user_item_matrix,k=5):
    similarity_score = user_similarity[user_id]
    similar_users=similarity_score.argsort()[::-1][1:k+1]
    recommended_books={}
    for similar_user in similar_users:
        similar_user_rating=user_item_matrix.iloc[similar_user]
        for isbn,rating in similar_user_rating.items():
            if rating>0 and user_item_matrix.loc[user_id,isbn]==0:
                if isbn in recommended_books:
                    recommended_books[isbn] += similarity_score[similar_user]*rating
                else:
                    recommended_books[isbn] = similarity_score[similar_user]*rating
    recommended_books=sorted(recommended_books.items(),key=lambda x:x[1],reverse=True)
    return recommended_books[:k]
```

```python
user_id=243
recommended_books=recommend(user_id,user_similarity,user_item_matrix,k=5)
print(f"Top 5 recommendation for 5 users {user_id}:")
for i,(isbn,score) in enumerate(recommended_books,1):
    book_title=df.loc[df['ISBN']==isbn,'Book-Title'].iloc[0]
    print(f'{i}.ISBN:{isbn},title:{book_title},score:{score}')
```
