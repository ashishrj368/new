```python
user_rating_df=rating_df.merge(user_df,left_on='User-ID',right_on='User-ID')
user_rating_df.head()
```

```python
book_df.head()
```

```python
book_user_rating_df=user_rating_df.merge(book_df,left_on='ISBN',right_on='ISBN')
book_user_rating_df.head()
```

```python
book_user_rating_df.isna().sum()
```

```python
book_user_rating_df=book_user_rating_df[['ISBN','Book-Title','Book-Author','User-ID','Book-Rating']]
book_user_rating_df.reset_index(drop=True,inplace=True)
book_user_rating_df.head()
```

```python
book_user_rating_df.isna().sum()
```

```python
df=book_user_rating_df[book_user_rating_df['Book-Rating']>=5]
df
```

```python
min_rating=book_user_rating_df['Book-Rating'].min()
max_rating=book_user_rating_df['Book-Rating'].max()
```

```python
book_user_rating_df['Book-Rating']=(book_user_rating_df['Book-Rating']-min_rating)/(max_rating-min_rating)*10
```

```python
book_user_rating_df.head()
```

```python
user_item_metrix=book_user_rating_df.pivot_table(index='User-ID',columns='ISBN',values='Book-Rating',fill_value=0)
```

```python
user_item_metrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similar_user=cosine_similarity(user_item_metrix)
similar_user
```

```python
def recommend(user_id,similar_user,user_item_metrix,k=5):
    user_similarity_score=similar_user[user_id]
    most_similar_users=user_similarity_score.argsort()[::-1][1:k+1]
    recommended_books={}
    for most_similar_user in most_similar_users:
        similar_user_rating=user_item_metrix.iloc[most_similar_user]
        for isbn,rating in similar_user_rating.items():
            if rating>0 and user_item_metrix.loc[user_id,isbn] ==0:
                if isbn in recommended_books:
                    recommended_books[isbn]+=user_similarity_score[most_similar_user]*rating
                else:
                    recommended_books[isbn] = user_similarity_score[most_similar_user]*rating
    recommended_books =sorted(recommended_books.items() ,key=lambda x:x[1] ,reverse=True)
    return recommended_books[:k]       
```

```python
user_id=487
recommended_books=recommend(user_id,similar_user,user_item_metrix,k=5)
print(f"Top 5 recommendation for users {user_id} :")
for i ,(isbn, score) in enumerate(recommended_books,1):
    book_title=df.loc[df['ISBN']==isbn, 'Book-Title'].iloc[0]
    print(f'{i} ISBN:{isbn} Book title:{book_title} score:{score}')
```
